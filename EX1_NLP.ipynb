# **a. Open the text file for processing**

```python
text_file = open("/content/nlp.txt")
text = text_file.read()
```

```python
print(type(text))
print("\n")
print(text)
print("\n")
print(len(text))
```

## **b. Import required libraries**

```python
from nltk import sent_tokenize
from nltk import word_tokenize
```

```python
import nltk
nltk.download('punkt')
```

## **c. Sentence Tokenizing**

```python
sentences = sent_tokenize(text)
print(len(sentences))
sentences
```

## **d. Word Tokenizing**

```python
words = word_tokenize(text)
print(len(words))
print(words)
```

## **e. Find the frequency distribution**

```python
from nltk.probability import FreqDist
fdist = FreqDist(words)
fdist.most_common(10)
```

## **f. Plot the frequency graph:**

```python
import matplotlib.pyplot as plt
fdist.plot(10)
```

## **g. Remove punctuation marks:**

```python
words_no_punc = []
for w in words:
  if w.isalpha():
    words_no_punc.append(w.lower())

print(words_no_punc)
print("\n")
print(len(words_no_punc))
```

## **h. Plotting graph without punctuation mark**

```python
fdist = FreqDist(words_no_punc)
fdist.most_common(10)
```

```python
fdist.plot(10)
```

## **i. List of stopwords**

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords = stopwords.words("english")
print(stopwords)
```

## **j. Removing stopwords**

```python
clean_words = []
for w in words_no_punc:
  if w not in stopwords:
    clean_words.append(w)

print(clean_words)
print("\n")
print(len(clean_words))
```

## **k. Final frequency distribution**

```python
fdist = FreqDist(clean_words)
fdist.most_common(10)
```

```python
fdist.plot(10)
```

## **Stemming**

## **a. Porter’s Stemmer Example 1**

```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()
word_list = ["Study", "Studying", "Studies", "Studied"]
for w in word_list:
  print(porter.stem(w))
```

## **b. Porter’s Stemmer Example 2**

```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()
word_list = ["Studies", "leaves", "decreases", "plays"]
for w in word_list:
  print(porter.stem(w))
```

## **c. SnowballStemmer**

```python
from nltk.stem import SnowballStemmer
snowball = SnowballStemmer("english")
word_list = ["Study", "Studying", "Studies", "Studied"]
for w in word_list:
  print(snowball.stem(w))
```

# **Lemmatizing**

# **a. A basic example demonstrating how a lemmatizer works**

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk import WordNetLemmatizer
lemma = WordNetLemmatizer()
word_list = ["Study", "Studying", "Studies", "Studied"]
for w in word_list:
  print(lemma.lemmatize(w, pos="v"))
```

# **b. Lemmatizer with default PoS value**

```python
from nltk import WordNetLemmatizer
lemma = WordNetLemmatizer()
word_list = ["Studies", "leaves", "decreases", "plays"]
for w in word_list:
  print(lemma.lemmatize(w))
```

# **c. Another example demonstrating the power of lemmatizer**

```python
from nltk import WordNetLemmatizer
lemma = WordNetLemmatizer()
word_list = ["am", "is", "are", "was", "were"]
for w in word_list:
  print(lemma.lemmatize(w, pos="v"))
```

# **d. Lemmatizer with different POS values - Parts Of Speech**

```python
from nltk import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize('studying', pos="v"))
print(lemmatizer.lemmatize('studying', pos="n"))
print(lemmatizer.lemmatize('studying', pos="a"))
print(lemmatizer.lemmatize('studying', pos="r"))
```
