## **FREQUENCY DISTRIBUTION**

```python
from textblob import TextBlob
import matplotlib.pyplot as plt

# Example text
# text = "This is an example sentence. Another sentence for demonstration purposes."

# Create a TextBlob object
blob = TextBlob(text)

# Get word frequencies
word_frequencies = blob.word_counts

# Sort the words and frequencies in descending order of frequencies
sorted_word_frequencies = sorted(word_frequencies.items(), key=lambda x: x[1], reverse=True)

# Extract words and their frequencies
words = [word for word, _ in sorted_word_frequencies]
frequencies = [freq for _, freq in sorted_word_frequencies]

# Plot the word frequencies as a line graph
plt.figure(figsize=(10, 6))
plt.plot(words, frequencies, marker='o')
plt.xticks(rotation=45)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Word Frequency Line Graph')
plt.tight_layout()
plt.show()
```

## **PUNCTUATION REMOVAL**

```python
from textblob import TextBlob

# Example text
# text = "Hello, world! How are you today?"

# Create a TextBlob object
blob = TextBlob(text)

# Remove punctuation from each word
filtered_text = " ".join([Word(word).strip().strip('.!?,:;"\'') for word in blob.words])

# Print the filtered text
print(filtered_text)
```

## **STOPWORDS REMOVAL**

```python
pip install textblob nltk
```

```python
import nltk
nltk.download('stopwords')
```

```python
from textblob import TextBlob
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Example text
# text = "This is an example sentence with stop words that should be removed."

# Tokenize the text
tokens = word_tokenize(text)

# Remove stop words from the tokens
filtered_tokens = [word for word in tokens if word.lower() not in stopwords.words('english')]

# Join the filtered tokens back into a string
filtered_text = " ".join(filtered_tokens)

# Print the filtered text
print(filtered_text)
```

## **FINAL FREQUENCY DISTRIBUTION**

```python
from textblob import TextBlob
from nltk import FreqDist
import matplotlib.pyplot as plt

# Example text
# text = "This is an example sentence. Another sentence for demonstration purposes."

# Create a TextBlob object
blob = TextBlob(text)

# Tokenize the text into words
words = blob.words

# Create a frequency distribution of words
freq_dist = FreqDist(words)

# Plot the frequency distribution
plt.figure(figsize=(10, 6))
freq_dist.plot(30, cumulative=False)
# plt.xlabel('Words')
# plt.ylabel('Frequency')
# plt.title('Word Frequency Distribution')
plt.show()
```

## **STEMMING**

```python
from textblob import TextBlob
from nltk.stem import PorterStemmer

# Example text
text = "This is an example sentence with running, runner, and runs."

# Create a TextBlob object
blob = TextBlob(text)

# Initialize Porter stemmer from NLTK
stemmer = PorterStemmer()

# Perform stemming on each word
stemmed_words = [stemmer.stem(word) for word in blob.words]

# Join the stemmed words into a sentence
stemmed_sentence = " ".join(stemmed_words)

# Print the stemmed sentence
print(stemmed_sentence)
```
